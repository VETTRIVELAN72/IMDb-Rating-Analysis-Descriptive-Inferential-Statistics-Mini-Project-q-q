```python
'''
Conclusions ->
(1) "imbd_votes" should be in integer format.
(2) "duration" needs to be converted into minutes (integer)
'''
```

```python
# Fix "imbd_votes" column by removing , and converting it to integer data type
df["imbd_votes"] = df["imbd_votes"].str.replace(",", "")
df["imbd_votes"] = pd.to_numeric(df["imbd_votes"], errors = 'coerce')
df.info()
```

```python
# Fix the duration : convert 'Xh Ym' format into minutes
def convert_duration(duration):
  if pd.isnull(duration):
    return None
  hours = 0
  minutes = 0
  if 'h' in duration:
    parts = duration.split('h')
    hours = int(parts[0].strip())
    if 'm' in parts[1]:
      minutes = int(parts[1].replace('m', '').strip())
  elif 'm' in duration:
    minutes = int(duration.replace('m', '').strip())
  return hours * 60 + minutes
```

```python
# Driver Code
df['duration_minutes'] = df['duration'].apply(convert_duration)
df.head()
```

```python
# Split the 'genre' column and OHE the extracted genres
genre_dummies = df['genre'].str.get_dummies(sep = ',')
df = pd.concat([df, genre_dummies], axis = 1)
df.head()
```

```python
# Drop unused columns
df_cleaned = df.drop(columns = ['title', 'genre', 'certificate', 'duration'])
df_cleaned.head()
```

```python
# Dropping NULL values
df_cleaned = df_cleaned.dropna()
df_cleaned.head()
```

```python
# Feature Engineering
```

```python
# Categorize year into ranges
def categorize_year(year):
  if 1990 <= year <= 2000:
    return '1990s'
  elif 2000 < year <= 2010:
    return '2000s'
  elif 2010 < year <= 2020:
    return '2010s'
  elif 2020 < year <= 2025:
    return '2020s'
  else:
    return 'Other'
```

```python
# Driver Code
df_cleaned['year_range'] = df_cleaned['year'].apply(categorize_year)
df_cleaned.head()
```

```python
# One Hot Encoding
year_dummies = pd.get_dummies(df_cleaned['year_range'], prefix = 'decade')
df_cleaned = pd.concat([df_cleaned, year_dummies], axis = 1)
df_cleaned.head()
```

```python
# Replacing False with 0 and True with 1 in the "df_cleaned" dataframe
df_cleaned = df_cleaned.replace({False: 0, True: 1})
df_cleaned.head()
```

```python
# Linear Regression Model ->
```

```python
# Seperating input and output features
X = df_cleaned[['imbd_votes', 'duration_minutes', 'Documentary', 'History', 'War']]
y = df_cleaned['imbd_rating']
```

```python
# We have taken only these input features because they have a strong relation in correlation heatmap.
```

```python
# Performing train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
# Creating the Linear Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Taking predictions from the model
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# Model Evaluation
from sklearn.metrics import r2_score
print('Training R2 Score :', np.round(r2_score(y_train, y_train_pred), 3))
print('Testing R2 Score :', np.round(r2_score(y_test, y_test_pred), 3))
```

```python
'''
As we can see both training and testing r2_score are way below expectation,
implying that the model is underfitted.
'''
```

```python
# Decision Tree Model ->
```

```python
# Seperating out numerical features
num_cols = df_cleaned.select_dtypes(include = np.number).columns
print(num_cols)
```

```python
# Removing "rank" and "imdb_rating" from list of numerical columns
num_cols = num_cols.drop(['rank', 'imbd_rating'])
print(num_cols)
```

```python
# Seperating numerical input features
X = df_cleaned[num_cols]
y = df_cleaned['imbd_rating']
```

```python
# Performing train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
# Creating a Decision Tree Model
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
# Taking the predictions from the model
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# Model Evaluation
from sklearn.metrics import r2_score
print('Training R2 Score :', np.round(r2_score(y_train, y_train_pred), 3))
print('Testing R2 Score :', np.round(r2_score(y_test, y_test_pred), 3))
```

```python
'''
As we can see, training R2 Score is perfect 1 and testing R2 Score is negative,
implying that the model is highly overfitted.
'''
```

```python
# Creating a parameter dictionary
param_grid = {
    'max_depth' : [5, 10, 20, 30],
    'min_samples_split' : [2, 5, 7, 10],
    'min_samples_leaf' : [1, 3, 5, 7, 9],
    'criterion' : ['squared_error', 'absolute_error', 'friedman_mse', 'poisson']

}
```

```python
# Using Grid Search CV to find optimal combination of hyper parameters
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = DecisionTreeRegressor(), param_grid = param_grid, cv = 5)
grid_search.fit(X_train, y_train)
```

```python
# Getting the best combination
print('Best Parameters :\n', grid_search.best_params_)
```

```python
# Creating an optimal model
model = DecisionTreeRegressor(criterion = 'squared_error', max_depth = 5,
                              min_samples_leaf = 9, min_samples_split = 10)
model.fit(X_train, y_train)
```

```python
# Taking the predictions from the model
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# Model Evaluation
from sklearn.metrics import r2_score
print('Training R2 Score :', np.round(r2_score(y_train, y_train_pred), 3))
print('Testing R2 Score :', np.round(r2_score(y_test, y_test_pred), 3))
```
